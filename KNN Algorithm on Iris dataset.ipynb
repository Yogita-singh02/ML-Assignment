# DATA CLEANING

```python
X=iris_data.drop(['species'],axis=1)
```

```python
y=iris_data.species
```

```python
# Perform test train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

# Training the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report , confusion_matrix
classifier=KNeighborsClassifier( p=2,metric='minkowski',n_neighbors=5)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred_train=classifier.predict(X_train)
y_pred_train
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_train, y_pred_train)
```

# Q2 Further train or fit into the model and calculate the performance for different values of k? (Use IRIS dataset from sk learn datasets)

```python
# Calculate the accuracy of the model for different values of k
for i in np.arange(1, 10):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    knn2.fit(X_train, y_train)
    print("For k = %d accuracy is"%i,knn2.score(X_test,y_test))
```
